Genere un nuevo dataset aplicando un filtro sobre los datos originales

```python
# Filtro por TV show 
Series = Netflix [Netflix ['type'] == "TV Show"]
Series.head (15)
```

```python
# Filtro por Country Colombia
Series = Netflix [Netflix ['country'] == "Colombia"]
Series.head (20)
```

```python
# Filtro multiple
## & : Conjuncion - AND
## | : Disyuncion - OR
# Filtro por peliculas colombianas lanzadas despues de 2020

series_co = Netflix[
    (Netflix["type"] == "Movie") &
    (Netflix["country"] == "Colombia") &
    (Netflix["release_year"] > 2020)
]
series_co.head ()
```

Genere un nuevo dataset aplicando un ordenamiento sobre los datos originales

```python
# Ordenamiento

print("Ordenado por año de lanzamiento (ascendienre):")
Netflix.sort_values(by=['release_year'], ascending=True)
```
